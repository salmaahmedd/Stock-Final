```python
import yfinance as yf
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# Fetch the Tesla ticker
tesla = yf.Ticker('TSLA')

# Fetch Tesla's stock price data
tesla_data = tesla.history(period="max")

# Define the function to create a graph
def make_graph(stock_data, company_name):
    # Create a single plot for stock price
    fig = go.Figure()

    # Add stock price trace
    fig.add_trace(
        go.Scatter(x=stock_data.index, y=stock_data['Close'], name=f"{company_name} Stock Price")
    )

    # Set titles and labels
    fig.update_layout(
        title=f"{company_name} Stock Price",
        xaxis_title="Date",
        yaxis_title="Stock Price"
    )

    # Show plot
    fig.show()

# Call the make_graph function
make_graph(tesla_data, "Tesla")
```
